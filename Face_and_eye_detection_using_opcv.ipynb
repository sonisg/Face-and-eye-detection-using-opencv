```
# Load the cascade
fac = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eyee= cv2.CascadeClassifier('haarcascade_eye.xml')
```

```
#Read input file
imge=cv2.imread("tesst.jpg") 
```

```
#convert into grey scale
if imge is not None: 
    grayy=cv2.cvtColor(imge,cv2.COLOR_BGR2GRAY)
    # ... other code ...
else:
    print("empty frame")
    exit(1)
```

```
# Detect faces
# Creating an object faces
faces= fac.detectMultiScale (grayy, 1.1, 10)
# Drawing rectangle around the face
for(x,y,w,h) in faces:
  cv2.rectangle(imge, (x,y) ,(x+w, y+h), (0,255,0), 3)
cv2_imshow(imge)  
```

```
# Creating two objects of interest
roi_gray=grayy[y:(y+h), x:(x+w)]
roi_color=imge[y:(y+h), x:(x+w)]
```

```
#for eyes detection
eyes = eyee.detectMultiScale(roi_gray, 1.1, 10)
for (x_eye, y_eye, w_eye, h_eye) in eyes:
  cv2.rectangle(roi_color,(x_eye, y_eye),(x_eye+w_eye, y_eye+h_eye), (0, 0, 255), 3) 
cv2_imshow(imge) 
```
